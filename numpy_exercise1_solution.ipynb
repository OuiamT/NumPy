```python
# 10
arr_ident = np.identity(3)
print(arr_ident)
# or
arr_eye = np.eye(3, 3, dtype=int)
print(arr_eye)
```

What is the difference between **eye()** and **identity()**?<br/>
In NumPy, **identity** creates a square identity matrix only, while **eye** creates an identity-like matrix of any shape.<br/>
In one line:
- np.identity(n) → always **n × n**
- np.eye(n, m) → **n × m** (more flexible)
